```python
import random
```

```python
#Function to decide each round
def ROLL(userin):
    compRoll = random.randint(0,2)
    print("\n")
    print("You threw {}".format(choices[int(userin)]))
    print("The computer threw {}".format(choices[compRoll]))
    if(int(userin) == compRoll):
        print("That was a draw")
    elif((userin == '0' and compRoll == 2) or (userin == '1' and compRoll == 0) or (userin == '2' and compRoll == 1)):
        print("AWESOME, you won this round")
        return True
    else:
        print("TOO BAD, you lost this round")
        return False

#Game START
print("Welcome to a game of ROCK PAPER & SCISSORS")
print("RULES: Rock beats Scissor, Paper beats Rock, Scissor beats Paper")
print("\n")
choices = ['Rock', 'Paper', 'Scissors']
#Ask user how many rounds to play
print("Decide on number of rounds: between 1 to 10")
while True:
    try:
        rounds = int(input('No. of Rounds: '))
    except ValueError:
        print("Please enter a number")
    if rounds < 1:
        print("Please enter a number between 1 to 10")
        continue
    elif rounds > 10:
        print("Please enter a number between 1 to 10")
        continue
    else:
        break
#set initial score to zero
userScore = 0
compScore = 0
#Loop to go around for the number of rounds picked by user
for x in range(0,rounds):
    while True:
        userRoll = input('Type your choice (rock, paper or scissor): ')
        if userRoll.casefold() not in ('rock','paper','scissor'):
            print("Not a valid choice")
            continue
        else:
            break
    if(userRoll.casefold() == 'rock'):
        userRoll = '0'
        poll = ROLL(userRoll)
        if(poll is True):
            userScore = userScore + 1
        elif(poll is False):
            compScore = compScore + 1
    elif(userRoll.casefold() == 'paper'):
        userRoll = '1'
        poll = ROLL(userRoll)
        if(poll is True):
            userScore = userScore + 1
        elif(poll is False):
            compScore = compScore + 1
    elif(userRoll.casefold() == 'scissor'):
        userRoll = '2'
        poll = ROLL(userRoll)
        if(poll is True):
            userScore = userScore + 1
        elif(poll is False):
            compScore = compScore + 1
    else:
        print('Invalid entry. Please type rock, paper or scissor')
    print('-------------------------------------------------------------------')
    print('\n')
    
#Displaying the final score
print("FINAL SCORE: ")
print("User Score: {}".format(userScore))
print("Computer Score: {}".format(compScore))
if(userScore > compScore):
    print("CONGRATULATIONS !!! YOU WON !!!!")
elif(userScore < compScore):
    print("SORRY LOSER !!!!!")
else:
    print("DAMN YOU LUCKY, ITS A DRAW !!!")
```
